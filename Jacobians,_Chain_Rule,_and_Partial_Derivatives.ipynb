```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Partial Derivatives

A **partial derivative** measures how a function changes with respect to one variable while keeping others constant.

Example function:

f(x, y) = x² + 3y²


```python
def f(x, y):
    return x**2 + 3*y**2

def partial_x(x, y):
    return 2*x

def partial_y(x, y):
    return 6*y

x, y = 2, 1
print("∂f/∂x:", partial_x(x, y))
print("∂f/∂y:", partial_y(x, y))
```

### Interpretation
- ∂f/∂x shows how the function changes if x changes.
- ∂f/∂y shows how the function changes if y changes.
- In ML, each model parameter has its own partial derivative.


## 2. Visualizing Partial Derivatives


```python
x_vals = np.linspace(-3, 3, 100)
plt.plot(x_vals, 2*x_vals)
plt.title("Partial Derivative with respect to x")
plt.xlabel("x")
plt.ylabel("2x")
plt.grid(True)
plt.show()
```

## 3. Chain Rule

The **Chain Rule** is used when functions are composed together.
Neural networks are repeated compositions of functions.

If:
z = g(y) and y = f(x)

Then:
dz/dx = dz/dy * dy/dx


```python
# Example chain rule
def f(x):
    return x**2

def g(y):
    return 3*y

x = 2
chain = 3 * (2*x)
print("Chain rule result:", chain)
```

### Why Chain Rule Matters in ML
Backpropagation in neural networks is entirely based on the chain rule.
Each layer passes gradients backward through multiplication.


## 4. Jacobian Matrix

For vector-valued functions, we use a **Jacobian matrix** instead of a simple derivative.

If F(x, y) → (u, v)

Jacobian =
[ ∂u/∂x  ∂u/∂y  
  ∂v/∂x  ∂v/∂y ]


```python
def jacobian(x, y):
    return np.array([[2*x, 0],
                     [0, 2*y]])

print(jacobian(1,2))
```

### Interpretation
- Jacobians describe how multiple outputs change with multiple inputs.
- Widely used in robotics, computer vision, and neural networks.


## 5. Practical ML Connection

- Partial derivatives → parameter updates  
- Chain rule → backpropagation  
- Jacobians → multi-output systems and transformations  

These tools allow gradient-based optimization to function correctly.


## Summary

- Partial derivatives measure change per variable.
- Chain rule enables gradient flow through layers.
- Jacobians generalize derivatives to vector outputs.
- These concepts form the mathematical backbone of neural network training.
